### Get the data from Yahoo Finance.

```python
import pandas as pd
import yfinance as yf # library to downloading historical stock prices
```

```python
df_yahoo = yf.download(['AAPL', 'MSFT'],
                      start='2000-01-01',
                      end='2020-10-05',
                      progress=False)
df_yahoo
```

```python
df_yahoo = yf.download('AAPL',
                      start='2000-01-01',
                      end='2020-10-05',
                      auto_adjust=True,
                      actions='inline', 
                      progress=False)
df_yahoo
```
