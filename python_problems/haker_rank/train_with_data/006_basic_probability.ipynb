## Event, Sample Space, and Probability

The probability of the occurrence of an event, $A$, is:
$$P(A) = \frac{Number of favorable outcomes}{Total number of outcomes}$$

Here are the first **two fundamental rules of probability**:\
>1) Any probbility, $P(A)$, is a number between 0 and 1 (i.e., $0 \le P(A) \le 1$)\
>2) The probability of the sample space, $S$, is $1$ (i.e., $P(S) = 1$)
    
**Sample space** is a just the set of the possible outcomes.

**ex.** The sample space is the set $S = {1, 2, 3, 4, 5, 6}$

A graphical representation of a sample space and events is a **Venn diagram**, as shown in Figure 3.1.1.\
In general the sample space S is represented by a rectangle, outcomes by points within the rectangle, and events\
by ovals that enclose the outcomes that compose them.



The outcomes that are even are  $2$,$4$, and $6$, so the event that corresponds to the phrase **an even number is rolled** is the set  ${2,4,6}$, which it is natural to denote by the letter  $E$.\
We write  $E=\{2,4,6\}$.

Similarly the event that corresponds to the phrase **a number greater than two is rolled** is the set  $T=\{3,4,5,6\}$, which we have denoted  $T$.\
>3) The probability that $P(A^c)$ will not occur, also equal that $P(A)$ is occur.\
$$P(A^c) = 1 - P(A)$$

**Notation**:\
$P(A^{'})$ - when we know that event will occur\
$P(A^{c})$ - when we know that event will not occur

## Question 1
---
Find the probability of getting an odd number when rolling a $6$-sided fair die.

Given the above question, we can extract the following:

Experiment: rolling a $6$-sided die.
Sample space ($S$): $S = \{1, 2, 3, 4, 5, 6\}$.\
Event ($A$): that the number rolled is odd (i.e., $A=\{1, 3, 5\}$).

If we refer back to the basic formula for the probability of the occurance fo an event, we can say:
$$P(A) = \frac{Number of favorable outcomes}{Total number of outcomes} = \frac{|A|}{|S|} = \frac{3}{6} = \frac{1}{2}$$

## Compound Events, Mutually Exclusive Events, and Collectively Exhaustive Events

**Compound Events** is a combination of $2$ or more simple events.

**Notation**:
Union () and Intersection () symbols LATEX is '\cap()' and the '\cup()' \
$\{ \}$ set a collection of elements - $A = \{3,7,9,14\}$, $B = \{9,14,28\}$

$A \cap B$ intersection - denotes the occurrence of $A$ and $B$ together.\
$A \cap B = \{9, 14\}$

$A \cup B$ union - denotes the occurrence of either $A$ or $B$.\
$A \cup B = \{3,7,9,14,28\}$

>4) A fundamental rule of probability: if $2$ events, $A$ and $B$, are disjoint, then the probability of either event is the sum of the probabilities of the $2$ events: $$P(A \ or \ B) = P(A) + P(B)$$

>5) The next a fundamental rule of probability: **the multiplication rule**.\
It states that if two events, $A$ and $B$, are independent, then the probability of both events is the product of the probabilities for each event: $$P(A \ \text{and} \ B) = P(A) * P(B)$$

## Question 2
---

Find the probability of getting $1$ head and $1$ tail when $2$ fair coins are tossed.

Given the above question, we can extract the following:\
Experiment: tossing $2$ coins.\
Sample space($S$): The possible outcomes for the toss of $1$ coin are $\{H, T\}$ where $H = \text{heads}$ and $T = \text{tails}$. As our experiment tosses $2$ coins, we have to consider all possible toss outcomes by finding the **Cartesian Product** of the possible outcomes for each coin: $S = \{\{H,T\} * \{H,T\} = \{(H,H),(H,T),(T,H),(T,T)\}$\
Event ($A \cap B$): That the outcome of $1$ toss will be $H$, and the outcome of the other toss will be $T$:
$$A = \{(H,T),(T,H)\}$$

Connecting this information back to our basic formula for $P(A)$, we can say:
$$P(A) = \frac{Number of favorable outcomes}{Total number of outcomes} = \frac{|A|}{|S|} = \frac{2}{4} = \frac{1}{2}$$

## Question 3
---

Let $A$ and $B$ be two events such that $P(A) = \frac{2}{5}$ and $P(B) = \frac{4}{5}$.\
If the probability of the occurence of either $A$ or $B$ is $\frac{3}{5}$, find the probability of the occurrence of both $A$ and $B$ together ($A \cap B$)

We can use our fundamental rules of probability to solve this problem:
$$|A \cup B| = |A| + |B| - |A \cap B|$$
$$\Rightarrow P(A \cup B) = P(A) + P(B) - P(A \cap B)$$
$$\Rightarrow P(A \cap B) = P(A) + P(B) - P(A \cup B)$$
$$\Rightarrow P(A \cap B) = \frac{2}{5} + \frac{4}{5} - \frac{3}{5} = \frac{3}{5}$$
